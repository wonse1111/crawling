```python
url_list = []
name_list = []
trans_name_list = []
amount_list = []
desc_list = []

# 각 상품의 정보에 각각 접근하기
# 순서대로 정보 저장하기

for sn in sn_info:
    url = sn.find('a').get('href')
    name = sn.find('p', {'class': 'name'}).text
    trans_name = sn.find('p', {'class': 'translated_name'}).text
    amount = sn.find('div', {'class': 'amount'}).text
    desc = sn.find('div', {'class': 'desc'}).text
    
    # 정보가 제대로 수집되었는지 확인
    
    print('이름 : {}, 가격 : {}원'.format(name,amount))
    
    url_list.append((base_url + url)) # base_url과 합쳐야 접근 가능한 주소 나옴
    name_list.append(name)
    trans_name_list.append(trans_name)
    amount_list.append(amount)
    desc_list.append(desc)

    
```

```python
# 데이터프레임에 저장

df = pd.DataFrame({'영문이름': name_list, '한글이름': trans_name_list, '가격': amount_list, '책정성격': desc_list, '링크': url_list})
df.head()
```
